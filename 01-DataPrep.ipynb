```python
import pandas as pd
import numpy as np
import pickle
```

# Load Data

There are two data set. The gene expression matrix (GEM) and physiological data taken at harvest (PhysData).

```python
GEM = pd.read_csv("01-InputData/GEMmaker.GEM.TMM.filtered.no.txt", low_memory=False, sep = "\t")
PhysData = pickle.load(open("01-InputData/fset_m2.pkl", "rb"))
GEM
```

```python
PhysData
```

# Data Preprocessing

Some samples do not contain starch values. These should be dropped from the dataset.

```python
# Setting the indices of the GEM to match the indices of the PhysData
GEM = GEM.T

GEM.index.name = PhysData.index.name
GEM
```

Removing samples that don't contain starch.

```python
PhysData = PhysData[PhysData['Starch_median'].notna()]
PhysData
```

Finding the common samples. Removing uncommon from GEM.

```python
commonIndices = PhysData.index.intersection(GEM.index)

print(len(commonIndices))
GEM = GEM.loc[commonIndices]
GEM
```

Removing uncommon samples from PhyData

```python
PhysData = PhysData.loc[commonIndices]
PhysData
```

# Pickle the Processed Data

```python
pickle.dump(GEM, open("GEM_final.pkl", "wb"))
pickle.dump(PhysData, open("PhysData.pkl", "wb"))
```
