## Build the Model

### Define Hyper-parameters

```python
input_size = 25210
hidden_size = 12605
output_size = 1
num_epochs = 100
learning_rate = 0.001
```

### Define the Model

```python
class StarchModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(StarchModel, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size, output_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, output_size)
        #self.dropout = nn.Dropout(p=0.25)
    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)
        return out

model = StarchModel(input_size, hidden_size, output_size).to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate, weight_decay=0.00001)
```
